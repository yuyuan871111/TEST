# interact 互動小應用

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

```python
def move(n):
    print(" "*n + "oooo")
```

```python
move(20)
```

```python
move(50)
```

```python
interact(move, n=(1,50))
```

## 圖型的互動應用 

```python
x = np.linspace(0, 2*np.pi, 1000)
x
```

```python
def draw(n):
    y = np.sin(n*x)
    plt.plot(x, y, lw=5)
    plt.show()
```

```python
draw(3)
```

```python
interact(draw, n=(1., 50.))
```

### 圖形的互動 over fitting

```python
from ipywidgets import interact_manual
```

```python
X = np.random.rand(6)
Y = np.random.rand(6)
```

```python
plt.scatter(X, Y, s=100, c='pink')
```

```python
x = np.linspace(0, 1, 1000)
```

```python
def draw(n):
    plt.scatter(X, Y, s=100, c='pink')
    y = 0.5*np.sin(n*x*2*np.pi) + 0.5
    plt.plot(x, y)
    plt.show()
```

```python
interact_manual(draw, n=(1., 200.))
```

## BMI 計算機

```python
def BMI_cal(height, weight):
    height = float(height)/100
    weight = float(weight)
    BMI = weight/(height**2)
    print("Your BMI is {:.2f}.".format(BMI))
```

```python
interact_manual(BMI_cal, height="FIll in your height(cm)", weight="Fill in your weight(kg)")
```
