## Pandas 的 loc 定位法

```python
df = pd.DataFrame(np.random.randn(5, 3),
                             index=list(range(1,6)),
                             columns=list("ABC"))
```

```python
df
```

```python
df[df.B>0]
```

```python
df[df.B>0]["C"]=0
```

```python
df
```

```python
df.loc[2:3, "B":"C"]
```

```python
df.loc[2, "B"]
```

```python
df.loc[2, "B"] =1
```

```python
df
```

```python
df.loc[df.B>0, "C"]=0
```

```python
df
```

```python
A =df["A"]
```

```python
A.plot()
```

### 報酬率

```python
r = A.diff()/A
```

```python
r[-3:].plot()
```

### 移動平均

```python
A_mean = A.rolling(window=2).mean()
```

```python
A_mean.plot()
A.plot()
A.rolling(window=3).mean().plot()
```
